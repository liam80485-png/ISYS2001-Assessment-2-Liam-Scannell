### Foundation Data Processing Functions

````python
def load_and_clean_transaction_data(file_path):
    required_columns = ['Date', 'Amount', 'Category', 'Description']

    try:
        if isinstance(file_path, str):
            df = pd.read_csv(file_path)
        else:
            df = pd.read_csv(file_path)

    except FileNotFoundError:
        raise FileNotFoundError(f"Business Error: The transaction file '{file_path}' was not found. Please ensure the correct path is provided.")
    except Exception as e:
        raise ValueError(f"Business Error: Could not load the transaction data. Please check the file format and content. Details: {e}")

    if not all(col in df.columns for col in required_columns):
        missing = [col for col in required_columns if col not in df.columns]
        raise ValueError(f"Business Error: The transaction data is missing essential columns: {', '.join(missing)}. Please ensure the CSV contains 'Date', 'Amount', 'Category', and 'Description'.")

    if 'Amount' in df.columns:
        try:
            df['Amount'] = df['Amount'].astype(str).str.replace(r'[$,€£]', '', regex=True).str.replace(',', '', regex=False)
            df['Amount'] = pd.to_numeric(df['Amount'], errors='coerce')
        except Exception as e:
            raise ValueError(f"Business Error: Failed to convert 'Amount' column to numbers. Please ensure amounts are in a valid numeric format (e.g., '$45.50' or '45.50'). Details: {e}")

    if 'Date' in df.columns:
        try:
            df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
        except Exception as e:
            raise ValueError(f"Business Error: Failed to convert 'Date' column to datetime format. Please ensure dates are in a recognizable format (e.g., 'YYYY-MM-DD'). Details: {e}")

    initial_rows = len(df)
    df.dropna(subset=['Date', 'Amount', 'Category', 'Description'], inplace=True)
    if len(df) < initial_rows:
        print(f"Warning: {initial_rows - len(df)} rows with missing or invalid essential data were removed.")

    return df
````

````python
def analyze_spending_patterns(df):
    if df.empty:
        return {
            "message": "No transactions to analyze.",
            "total_spending": 0.0,
            "category_spending": {},
            "spending_percentages": {},
            "top_category": "N/A",
            "average_transaction_amount": 0.0,
            "insights": []
        }

    total_spending = df['Amount'].sum()

    category_spending = df.groupby('Category')['Amount'].sum().sort_values(ascending=False)

    positive_spending = df[df['Amount'] > 0]['Amount'].sum()
    spending_percentages = {}
    if positive_spending > 0:
        for category, amount in category_spending.items():
            if amount > 0:
                spending_percentages[category] = (amount / positive_spending) * 100
            else:
                spending_percentages[category] = 0.0

    spending_percentages = {k: v for k, v in sorted(spending_percentages.items(), key=lambda item: item[1], reverse=True)}

    top_category = category_spending.index[0] if not category_spending.empty else "N/A"

    positive_transactions = df[df['Amount'] > 0]['Amount']
    average_transaction_amount = positive_transactions.mean() if not positive_transactions.empty else 0.0

    insights = []
    if total_spending > 0:
        insights.append(f"Your total expenditure across all categories is ${total_spending:.2f}.")
        insights.append(f"The highest spending category is '{top_category}' accounting for ${category_spending.iloc[0]:.2f}.")
        if 'Flights' in category_spending and 'Accommodation' in category_spending:
            insights.append("Flights and Accommodation are typically the largest components of travel budgets. Consider booking in advance or looking for package deals to save.")
        elif 'Dining' in category_spending and 'Activities' in category_spending:
            insights.append("Significant spending on Dining and Activities suggests a focus on experiences. Look for local deals or cooking some meals to reduce costs.")
        if average_transaction_amount > 100:
            insights.append(f"Your average positive transaction amount is ${average_transaction_amount:.2f}. Monitoring larger individual expenses could help with budget management.")
    elif total_spending < 0:
        insights.append(f"Your net balance is positive due to refunds, with a total of ${total_spending:.2f}. Good job on managing returns!")
    else:
        insights.append("Your net spending is zero, indicating a balanced budget or transactions primarily consisting of refunds.")


    return {
        "message": "Spending analysis completed successfully.",
        "total_spending": total_spending,
        "category_spending": category_spending.to_dict(),
        "spending_percentages": spending_percentages,
        "top_category": top_category,
        "average_transaction_amount": average_transaction_amount,
        "insights": insights
    }

from io import StringIO
import pandas as pd

sample_transactions = {
    'Date': ['2024-12-20', '2024-12-20', '2024-12-21', '2024-12-21', '2024-12-22', '2024-12-22', '2024-12-23', '2024-12-23', '2024-12-24', '2024-12-24', '2024-12-25', '2024-12-25', '2024-12-26', '2024-12-26', '2024-12-27'],
    'Amount': ['$850.00', '$120.50', '$250.00', '$45.75', '$75.00', '$30.00', '$150.00', '$60.00', '$95.20', '$22.50', '$400.00', '$88.00', '$180.00', '$15.00', '$50.00'],
    'Category': ['Flights', 'Dining', 'Accommodation', 'Transport', 'Activities', 'Souvenirs', 'Dining', 'Shopping', 'Activities', 'Coffee', 'Accommodation', 'Dining', 'Flights', 'Transport', 'Groceries'],
    'Description': ['Flight to Bali', 'Airport Dinner - Qantas Lounge', 'Hotel Stay - Seminyak', 'Taxi to Hotel', 'Surf Lesson', 'Local Craft Market', 'Seafood BBQ', 'Beachwear Shopping', 'Snorkeling Tour', 'Beach Cafe Latte', 'Return Flight to Sydney', 'Farewell Dinner', 'Hotel Check-out', 'Souvenir Shopping', 'Airport Shuttle']
}
df_sample = pd.DataFrame(sample_transactions)

sample_csv_data_for_test = df_sample.to_csv(index=False)
test_file_path_for_analysis = StringIO(sample_csv_data_for_test)

try:
    cleaned_df_for_analysis = load_and_clean_transaction_data(test_file_path_for_analysis)
    analysis = analyze_spending_patterns(cleaned_df_for_analysis)
    print("\n✅ Spending analysis successful!")
    print("Total Spending: ", analysis['total_spending'])
    print("Category Spending: ", analysis['category_spending'])
    print("Spending Percentages: ", analysis['spending_percentages'])
    print("Top Category: ", analysis['top_category'])
    print("Average Transaction Amount: ", analysis['average_transaction_amount'])
    print("Insights:")
    for insight in analysis['insights']:
        print(f"- {insight}")
except Exception as e:
    print(f"\n❌ Error during spending analysis test: {e}")
````

````python
def generate_financial_recommendations(analysis_data):
    report_sections = []

    total_spending = analysis_data.get('total_spending', 0.0)
    message = analysis_data.get('message', 'No specific message.')
    report_sections.append(f"--- Financial Recommendations Report ---")
    report_sections.append(f"Status: {message}")
    report_sections.append(f"Overall Total Spending: ${total_spending:.2f}\n")

    category_spending = analysis_data.get('category_spending', {})
    spending_percentages = analysis_data.get('spending_percentages', {})
    top_category = analysis_data.get('top_category', 'N/A')
    average_transaction_amount = analysis_data.get('average_transaction_amount', 0.0)

    report_sections.append("1. Spending Pattern Observations:")
    if category_spending:
        report_sections.append(f"   - Your highest spending category is **{top_category}** at ${category_spending.get(top_category, 0.0):.2f}. This accounts for {spending_percentages.get(top_category, 0.0):.2f}% of your positive spending.")
        if top_category in ['Flights', 'Accommodation']:
            report_sections.append("     Recommendation: For large travel expenses like flights and accommodation, consider booking further in advance, looking for package deals, or exploring loyalty programs to reduce costs.")
        elif top_category == 'Dining':
            report_sections.append("     Recommendation: Dining is a common area for budget optimization. Try meal prepping, exploring budget-friendly local eateries, or taking advantage of happy hour deals.")
        elif top_category == 'Shopping':
            report_sections.append("     Recommendation: Review your shopping habits. Differentiate between needs and wants, and consider a '24-hour rule' for non-essential purchases to avoid impulse buying.")
        else:
            report_sections.append(f"     Observation: Further review of '{top_category}' spending could reveal specific areas for optimization.")

        if average_transaction_amount > 100:
            report_sections.append(f"   - Your average positive transaction is ${average_transaction_amount:.2f}. Monitoring these larger individual expenses can help manage overall budget.")
    else:
        report_sections.append("   - No detailed category spending to analyze.")

    report_sections.append("\n2. Potential Savings Opportunities:")
    if 'Flights' in category_spending or 'Accommodation' in category_spending:
        report_sections.append("   - **Travel Costs:** Given your spending on travel-related categories, significant savings could be achieved by comparing prices across different platforms, being flexible with dates, or using points/miles if applicable.")
    if 'Dining' in category_spending and spending_percentages.get('Dining', 0) > 10:
        report_sections.append("   - **Dining Out:** High dining expenses indicate an opportunity to save. Cooking at home more often, packing lunches, or limiting restaurant visits to special occasions can significantly reduce this category.")
    if 'Coffee' in category_spending and category_spending['Coffee'] > 20:
         report_sections.append("   - **Small Frequent Purchases:** Even small daily purchases like coffee can add up. Consider making coffee at home a few days a week.")

    initial_insights = analysis_data.get('insights', [])
    if initial_insights:
        report_sections.append("\n3. Additional Insights:")
        for insight in initial_insights:
            report_sections.append(f"   - {insight}")

    report_sections.append("\n--- Actionable Advice ---")
    report_sections.append("\n*   **Set a Budget:** Establish clear spending limits for your top categories like 'Flights' and 'Accommodation' to prevent overspending.")
    report_sections.append("*   **Track Expenses:** Continue tracking your expenses diligently to monitor progress and identify any new patterns.")
    report_sections.append("*   **Review Regularly:** Schedule a weekly or monthly review of your spending to stay on top of your financial goals.")
    report_sections.append("*   **Seek Alternatives:** For categories with high spending, actively look for more cost-effective alternatives.")

    return "\n".join(report_sections)
````

````python
import pandas as pd

financial_documents = [
    {
        "id": "travel_budgeting",
        "content": (
            "Travellers should divide expenses into flights, "
            "accommodation, dining, transport, shopping, and activities."
        )
    },
    {
        "id": "flight_savings",
        "content": (
            "Booking flights early and travelling during off-peak "
            "seasons can reduce travel costs."
        )
    },
    {
        "id": "accommodation_savings",
        "content": (
            "Budget hotels, hostels, and shared accommodation "
            "can help reduce accommodation expenses."
        )
    },
    {
        "id": "currency_conversion",
        "content": (
            "Currency conversion helps travellers compare spending "
            "between different countries and currencies."
        )
    }
]

def retrieve_documents(user_question):
    results = []
    question = user_question.lower()
    for doc in financial_documents:
        content = doc["content"].lower()
        if (
            "flight" in question and "flight" in content
        ) or (
            "accommodation" in question and "accommodation" in content
        ) or (
            "currency" in question and "currency" in content
        ):
            results.append(doc["content"])
    return results

print("Testing Simple RAG System...\n")

question = (
    "How can I reduce spending on flights "
    "and accommodation?"
)

print(f"User Question: {question}\n")

documents = retrieve_documents(question)

print("Retrieved Documents:\n")

for doc in documents:
    print(f"- {doc}\n")
````

## Custom Financial Tools

````python
class FinanceAgent:

    def __init__(self):
        self.tools = {}

    def register_tool(self, tool_name, tool_function):
        self.tools[tool_name] = tool_function
        print(f"✅ Tool '{tool_name}' registered successfully!")

    def use_tool(self, tool_name, *args):
        if tool_name in self.tools:
            return self.tools[tool_name](*args)
        else:
            return "❌ Tool not found."


finance_agent = FinanceAgent()

def savings_calculator(
    current_savings,
    monthly_contribution,
    target_amount
):
    if (
        current_savings < 0 or
        monthly_contribution <= 0 or
        target_amount <= 0
    ):
        return "❌ Invalid input values."

    if current_savings >= target_amount:
        return (
            f"✅ Savings goal already reached!"
        )

    remaining = (
        target_amount - current_savings
    )

    months_needed = (
        remaining / monthly_contribution
    )

    years = int(months_needed // 12)

    months = int(months_needed % 12)

    return (
        f"It will take approximately "
        f"{years} year(s) and "
        f"{months} month(s) "
        f"to reach your savings goal."
    )


finance_agent.register_tool(
    "savings_calculator",
    savings_calculator
)


result = finance_agent.use_tool(
    "savings_calculator",
    1000,
    200,
    5000
)

print("\nSavings Calculator Result:")
print(result)
````
